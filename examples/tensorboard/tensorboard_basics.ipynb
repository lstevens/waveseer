## 2. Understanding the TensorBoard Interface

TensorBoard provides several dashboards:

- **Scalars**: Track metrics like loss and accuracy over time
- **Graphs**: Visualize model architecture
- **Distributions**: Monitor weight and gradient distributions
- **Histograms**: Track how distributions change over time
- **Images**: View sample images or visualizations
- **Text**: Log text data like descriptions or outputs
- **PR Curves**: View precision-recall curves
- **Mesh**: View 3D meshes
- **HParams**: View hyperparameter tuning results

## 3. Creating a Simple TensorBoard Example

Let's create a simple example to demonstrate TensorBoard basics:

````python
# Create a log directory
log_dir = "../runs/tensorboard_demo_" + datetime.now().strftime("%Y%m%d_%H%M%S")
writer = SummaryWriter(log_dir=log_dir)

# Generate dummy data
for i in range(100):
    # Log scalar values (like loss)
    writer.add_scalar('Loss/train', np.random.random() * np.exp(-i/30), i)
    writer.add_scalar('Loss/validation', np.random.random() * np.exp(-i/30) + 0.05, i)
    
    # Log accuracy
    writer.add_scalar('Accuracy/train', 1 - np.random.random() * np.exp(-i/30), i)
    writer.add_scalar('Accuracy/validation', 1 - (np.random.random() * np.exp(-i/30) + 0.05), i)
    
    # Log learning rate
    writer.add_scalar('Learning_rate', 0.001 * np.exp(-i/50), i)

writer.close()

print(f"TensorBoard logs written to {log_dir}")
print("Refresh the TensorBoard interface to see the new data.")
````

## 4. Visualizing Model Architecture

TensorBoard can visualize your model architecture, which is helpful for understanding complex models:

````python
# Create a simple CNN model
model = CNNPatternModel(n_features=1, n_classes=3, hidden_size=64, kernel_sizes=[3, 5, 7], channels=[32, 64, 128], dropout=0.2)

# Create a writer for the model
model_log_dir = "../runs/model_viz_" + datetime.now().strftime("%Y%m%d_%H%M%S")
model_writer = SummaryWriter(log_dir=model_log_dir)

# Create a sample input tensor (batch_size, channels, sequence_length)
sample_input = torch.randn(1, 1, 100)

# Add the model graph to TensorBoard
try:
    model_writer.add_graph(model, sample_input)
except Exception as e:
    print(f"Could not add graph to TensorBoard: {e}")
    print("This is normal if the model contains operations not supported by TensorBoard.")

model_writer.close()

print(f"Model visualization logs written to {model_log_dir}")
print("Refresh the TensorBoard interface and go to the 'Graphs' tab to see the model architecture.")
````

## 5. Using the Waveseer TensorBoard Utilities

The Waveseer library provides utilities to make TensorBoard integration easier:

````python
# Create a TensorBoard writer using the utility function
writer = setup_tensorboard("demo_model")

# Create sample data
n_samples = 100
sequence_length = 50
n_features = 1

# Synthetic data
X = torch.randn(n_samples, n_features, sequence_length)
y = torch.randint(0, 3, (n_samples,))  # 3 classes

# Create a simple LSTM model
lstm_model = LSTMPatternModel(
    n_features=n_features, 
    n_classes=3, 
    hidden_size=64, 
    num_layers=2, 
    dropout=0.2,
    bidirectional=True
)

# Log the model graph
log_model_graph(writer, lstm_model, X[0:1])  # Add a single sample

# Close the writer
writer.close()

print(f"LSTM model graph added to {writer.log_dir}")
print("Refresh TensorBoard to see the LSTM model architecture.")
````
